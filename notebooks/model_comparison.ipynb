```python
# STEP 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# STEP 2: Load Results from CSV files

rf = pd.read_csv("../results/random_forest.csv")
lr = pd.read_csv("../results/logistic_regression.csv")
svm = pd.read_csv("../results/svm.csv")
dt = pd.read_csv("../results/decision_tree.csv")

# Combine all results
df = pd.concat([lr, svm, dt, rf], ignore_index=True)

print(df)
```

```python
# STEP 3: Display Table Nicely

df.set_index("Model", inplace=True)
df
```

```python
# STEP 4: Accuracy Comparison (Bar Chart)

plt.figure()
df["Accuracy"].plot(kind='bar')
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.show()
```

```python
# STEP 5: F1 Score Comparison

plt.figure()
df["F1-score"].plot(kind='bar')
plt.title("Model F1 Score Comparison")
plt.ylabel("F1 Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# STEP 6: ROC AUC Comparison

plt.figure()
df["ROC_AUC"].plot(kind='bar')
plt.title("Model ROC AUC Comparison")
plt.ylabel("ROC AUC")
plt.xticks(rotation=45)
plt.show()
```

```python
# STEP 7: Combined Comparison Plot

df.plot(kind='bar')
plt.title("Overall Model Comparison")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.show()
```

```python
# STEP 8: Highlight Best Model

best_model = df["Accuracy"].idxmax()
print("Best Model based on Accuracy:", best_model)
```

```python
# STEP 9: Simple Ranking

df["Rank"] = df["Accuracy"].rank(ascending=False)
print(df.sort_values("Rank"))
```
