```python
# 1. Load Data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/online_gaming_behavior_dataset.csv")

print(df.shape)
df.head()
```

```python
# 2. Target Distribution
df['EngagementLevel'].value_counts().plot(kind='bar')

plt.title("Engagement Level Distribution")
plt.xlabel("Engagement Level")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Numerical Feature Distribution
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplot (Outliers)
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=df[['PlayTimeHours','SessionsPerWeek','AvgSessionDurationMinutes']])
plt.title("Boxplot of Key Features")
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(10,8))

corr = df.select_dtypes(include=['int64','float64']).corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# 6. Categorical feature distribution

categorical_cols = ['Gender', 'GameGenre', 'GameDifficulty']

for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```
